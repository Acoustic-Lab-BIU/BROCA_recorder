```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
SHEET_PATH = 'NLU Test Set for BIU.xlsx'
SAVE_DIR = Path('sentence_sheets')
sentence_excel = pd.read_excel(SHEET_PATH,sheet_name='FINAL SET')
sentence_excel
```

```python
shuffled_df = sentence_excel.sample(frac=1).reset_index(drop=True)
shuffled_df
```

```python
for i in range(1,6):
    df = shuffled_df.loc[(i-1)*20:i*20-1].copy().reset_index(drop=True)
    df.to_csv(SAVE_DIR/f'session{i}.csv',index=False)
```
